```python
# Function to generate Pascal's Triangle up to n rows
def generate_pascals_triangle(n):
    triangle = []  # List to hold all rows of Pascal's Triangle
    
    for i in range(n):
        row = [1]  # Every row starts with 1
        if i > 0:
            for j in range(1, i):  # Calculate the intermediate values of the row
                # Add the two values from the previous row to get the current value
                row.append(triangle[i-1][j-1] + triangle[i-1][j])
            row.append(1)  # Every row ends with 1
        triangle.append(row)  # Add the row to the triangle
    
    return triangle

# Function to print Pascal's Triangle with tab-separated values
def print_pascals_triangle(triangle):
    for row in triangle:
        print("\t".join(map(str, row)))  # Convert each row to string and join with tabs

# Generate the first 8 rows of Pascal's Triangle
rows = 8
pascals_triangle = generate_pascals_triangle(rows)

# Print Pascal's Triangle
print_pascals_triangle(pascals_triangle)
```

```python
import random

def generate_password(length):
    # Ensure that the length is valid
    if length <= 0:
        return "Invalid length. Please enter a positive number."

    password = ""
    for _ in range(length):
        # Generate a random number between 33 and 126 (inclusive)
        random_decimal = random.randint(33, 126)
        # Convert the random number to the corresponding ASCII character
        password += chr(random_decimal)
    
    return password

# Example: Generate a password of length 12
password_length = 12
password = generate_password(password_length)
print("Generated Password:", password)
```
